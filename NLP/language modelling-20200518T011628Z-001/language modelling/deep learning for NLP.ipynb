```
from nltk.corpus import stopwords
import string, re

def to_token(data):
  data_split = data.split()
  #prepare regex for char filtering
  re_punc = re.compile('[%s]' %re.escape(string.punctuation))
  #remove punctuation from each word
  tokens = [re_punc.sub('', w) for w in data_split]
  #remove remainingg tokens that are not alphabetic
  tokens = [word for word in tokens if word.isalpha()]
  #filter out stop words
  stop_words = set(stopwords.words('english'))
  tokens = [w for w in tokens if w not in stop_words]
  #filter out short tokens
  tokens = [word for word in tokens if len(word) >1]


  return tokens

data['transformed_tokens'] = data['text'].apply(to_token)
```

```
from collections import Counter

def add_doc_to_vocab(filename, vocab):
  #load doc
  doc = load_doc(filename)
  #clean doc
  tokens = to_token(doc)
  # update counts
  vocab.update(tokens)

def process_docs(directory, vocab):
  for filename in os.listdir(directory):
    if not filename.endswith(('.txt')):
      next
    path = directory + '/'+filename

    add_doc_to_vocab(path, vocab)

#define vocab
vocab = Counter()
process_docs('neg/neg', vocab)
process_docs('pos', vocab)
print(len(vocab))
print(vocab.most_common(50))
```

```
# comparing word scoring methods
def prepare_data(train_docs, test_docs, mode):
  #create the tokenizer
  tokenizer = Tokenizer(num_words = 100)
  tokenizer.fit_on_texts(train_docs)
  xtrain = tokenizer.texts_to_matrix(train_docs, mode = mode)
  xtest = tokenizer.texts_to_matrix(test_docs, mode = mode)
  return xtrain, xtest

def evaluate_mode(x_train, y_train, x_test, y_test):
  scores = list()
  n_repoeats = 30
  n_words = xtest.shape[1]
  for i in range(n_repeats):
    #define model
    model = Sequential()
    model.add(Dense(50, input_shape = (n_words, ), activaton = 'relu'))
    model.add(Dense(1, activation = 'sigmoid'))
    #compile network
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    #fit network
    model.fit(x_train, y_train, epochs = 10, verbose = 2)
    #evaluate
    loss, acc = model.evaluate(x_test, y_test, verbose = 0)
    scores.append(acc)
    print('%d accuracy: %s' %((i+1), acc))
  return scores

```
